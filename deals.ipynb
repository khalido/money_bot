```python
def get_deals(q="mobile", num_deals=3, data=data_list):
    """takes in a category q and returns the top deals"""
    url = "https://www.ozbargain.com.au/cat/" + q +"/feed"
    
    payload = {"query": query.lower()}
    
    r = requests.get(url)
    if r.status_code is not requests.codes.ok:
        # try the tags deals sections
        url = "https://www.ozbargain.com.au/tag/" + q +"/feed"
        r = requests.get(url)
        if r.status_code is not requests.codes.ok:
              # send a msg saying deal not found
                print(f"no deals found for {q}, pls try something else")
                return
        
    deals = feedparser.parse(r.text)['entries']
    
    for i in range(num_deals):
        title = deals[i]['title'][:80]
        link = deals[i]['link']
        image_url = deals[i]['media_thumbnail'][0]['url']
        
        # send msg to user
        print(title,link, len(title))
        if i == 0:
            data['message']['attachment']['payload']['elements'][i]['title'] = title
            data['message']['attachment']['payload']['elements'][i]['default_action']['url'] = link
            data['message']['attachment']['payload']['elements'][i]['default_action']['default_action'] = link
            data['message']['attachment']['payload']['elements'][i]['default_action']['fallback_url'] = link
        else:
            data['message']['attachment']['payload']['elements'][i]['default_action']
            data['message']['attachment']['payload']['elements'][i]['title'] = title
            data['message']['attachment']['payload']['elements'][i]['image_url'] = image_url
            
    print("----"*4)
    print(data)

get_deals("computing")
```
